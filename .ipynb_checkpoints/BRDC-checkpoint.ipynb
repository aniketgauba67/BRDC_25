```python
# Sam's Fans Fundraising Data Analysis Notebook

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
# Set styles
sns.set(style="whitegrid")
```

```python
# 2. Load Data
# Updated to load CSV file
# Ensure Donation_Data.csv is in the same directory

data = pd.read_csv('Donar_Data.csv', dtype={'Date': str})

data.head()
```

```python
# 3. Quick Data Check
data.info()
data.describe()
```

```python
# 4. Clean and Prepare Data
# Assume columns like: 'Donor ID', 'Amount', 'Date', 'Campaign', 'Recurring Payment?'


# 2. Parse Date column correctly (flexible format)
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# 3. Check first 10 dates
print(data['Date'].head(10))

# 4. Now create YearMonth
data['YearMonth'] = data['Date'].dt.to_period('M')
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
```

```python
# 1. Clean Revenue Amount properly
data['Revenue Amount'] = data['Revenue Amount'].replace(r'[\$,]', '', regex=True).astype(float)

# 2. Simplify donation types
def categorize_type(x):
    if 'Recurring' in str(x):
        return 'Recurring'
    else:
        return 'One-Time'

data['Donation Type'] = data['Type'].apply(categorize_type)

# 3. Group
recurring_summary = data.groupby('Donation Type')['Revenue Amount'].sum()

# 4. Plot (visually enhanced)
fig, ax = plt.subplots(figsize=(8,6))

bars = ax.bar(recurring_summary.index, 
               recurring_summary.values, 
               color=['#4CAF50', '#2196F3'],  # Green and Blue
               edgecolor='black')

# Title and labels
ax.set_title('Recurring vs One-Time Donations', fontsize=16, fontweight='bold')
ax.set_xlabel('Donation Type', fontsize=14)
ax.set_ylabel('Total Donation Amount ($)', fontsize=14)

# Y-axis in thousands (e3)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x*1e-3)}k'))

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height/1000:.1f}k', 
                xy=(bar.get_x() + bar.get_width() / 2, height), 
                xytext=(0, 5), 
                textcoords="offset points",
                ha='center', va='bottom', fontsize=12, fontweight='bold')

# Make x-axis labels horizontal
plt.xticks(rotation=0, fontsize=12)

# Add grid
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('your_plot_name.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Group monthly donations
monthly_donations = data.groupby('YearMonth')['Revenue Amount'].sum().reset_index()

# 2. Calculate Moving Average (6-month rolling)
monthly_donations['6_Month_MA'] = monthly_donations['Revenue Amount'].rolling(window=6, min_periods=1).mean()

# 3. Plot
fig, ax = plt.subplots(figsize=(14, 7))

# Main donations line
ax.plot(monthly_donations['YearMonth'].astype(str), 
        monthly_donations['Revenue Amount'], 
        marker='o', 
        linestyle='-', 
        color='#FF6F61', 
        linewidth=2, 
        label='Monthly Donations')

# Moving Average line
ax.plot(monthly_donations['YearMonth'].astype(str), 
        monthly_donations['6_Month_MA'], 
        color='#005f73', 
        linewidth=3, 
        linestyle='--', 
        label='6-Month Moving Average')

# Title and labels
ax.set_title('Total Donations Over Time (with Moving Average)', fontsize=20, fontweight='bold')
ax.set_xlabel('Month-Year', fontsize=16)
ax.set_ylabel('Total Donations ($)', fontsize=16)

# Format Y-axis in Thousands
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x*1e-3)}k'))

# Fix X-axis: show label every 6 months
step = 6
plt.xticks(ticks=range(0, len(monthly_donations), step), 
           labels=monthly_donations['YearMonth'].astype(str)[::step], 
           rotation=45, ha='right', fontsize=10)

# Grid
ax.grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# OPTIONAL: Annotate only very high spikes
threshold = monthly_donations['Revenue Amount'].quantile(0.95)
for i, row in monthly_donations.iterrows():
    if row['Revenue Amount'] >= threshold:
        ax.annotate(f'{row['Revenue Amount']/1000:.1f}k', 
                    (i, row['Revenue Amount']), 
                    textcoords="offset points", 
                    xytext=(0,10), 
                    ha='center', fontsize=8, fontweight='bold', color='black')

# Add legend
ax.legend(fontsize=12)

# Layout
plt.tight_layout()

# Show plot
plt.savefig('your_plot_name.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

def plot_campaign_vs_unprompted(data):
    """
    Plots a professional bar chart comparing Campaign vs Unprompted Donations.
    Automatically handles missing/empty campaign names.
    """
    # Step 1: Create clean Campaign_Flag
    data = data.copy()
    if 'Campaign' not in data.columns:
        raise ValueError("The dataset must have a 'Campaign' column.")

    data['Campaign_Flag'] = data['Campaign'].apply(lambda x: 'Campaign' if pd.notna(x) and str(x).strip() != '' else 'Unprompted')

    # Step 2: Group data
    campaign_summary = data.groupby('Campaign_Flag')['Revenue Amount'].sum().sort_values(ascending=False)

    if campaign_summary.empty:
        print(" No Campaign or Unprompted data available to plot.")
        return

    # Step 3: Setup plot
    fig, ax = plt.subplots(figsize=(10, 6))
    colors = ['#48cae4', '#0077b6']

    # Step 4: Bar Plot
    bars = ax.bar(campaign_summary.index, 
                  campaign_summary.values, 
                  color=colors[:len(campaign_summary)],  # Auto handle color length
                  edgecolor='black', 
                  width=0.6)

    # Step 5: Big Bold Title
    plt.suptitle('Campaigns Drive Higher Donations', fontsize=22, fontweight='bold', color='#03045e')
    plt.title('Boosting Engagement Through Focused Fundraising Efforts', fontsize=16, color='#555', pad=8)

    # Step 6: Labels
    ax.set_xlabel('Donation Source', fontsize=14)
    ax.set_ylabel('Total Donations ($)', fontsize=14)

    # Step 7: Y-axis formatting
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'${int(x/1000)}k'))

    # Step 8: Annotate the bars
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height/1000:.1f}k', 
                    xy=(bar.get_x() + bar.get_width()/2, height), 
                    xytext=(0,5), 
                    textcoords='offset points',
                    ha='center', va='bottom', fontsize=12, fontweight='bold', color='navy')

    # Step 9: Light highlight box takeaway
    plt.text(0.5, -0.2, 
             'Fundraising campaigns generate over **5x** more revenue than unprompted giving.', 
             fontsize=12, ha='center', color='#03045e', 
             transform=ax.transAxes, bbox=dict(facecolor='#caf0f8', edgecolor='#00b4d8', boxstyle='round,pad=0.5'))

    # Step 10: Style tweaks
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    fig.patch.set_facecolor('#f7f9fb')

    plt.tight_layout()
    plt.show()
data['Campaign'] = data['Campaign'].str.strip()
plot_campaign_vs_unprompted(data)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sort and prepare data
donations = data['Revenue Amount']
donations = donations[donations > 0].sort_values(ascending=True).reset_index(drop=True)

# Cumulative revenue
cumulative_revenue = donations.cumsum()
total_revenue = cumulative_revenue.iloc[-1]
cumulative_percentage = cumulative_revenue / total_revenue

# Find 80% point
threshold_idx = cumulative_percentage[cumulative_percentage >= 0.8].index[0]
threshold_value = donations.iloc[threshold_idx]

# Plot
fig, ax = plt.subplots(figsize=(12, 7))

# Main cumulative curve
ax.plot(donations, cumulative_percentage, color='seagreen', linewidth=3)

# Highlight 80% zone
ax.axhline(0.8, color='gray', linestyle='--', linewidth=1.5)
ax.axvline(threshold_value, color='red', linestyle='--', linewidth=2)

# Fill area under 80%
ax.fill_between(donations, 0, cumulative_percentage, 
                where=(cumulative_percentage <= 0.8), 
                color='lightgreen', alpha=0.5, label='Top Revenue Zone')

# Annotations
ax.text(threshold_value, 0.05, f'80% revenue at\n${int(threshold_value):,}', 
        color='red', fontsize=12, rotation=90, verticalalignment='bottom', fontweight='bold')

ax.annotate('Major Gift Zone', xy=(threshold_value, 0.8), xytext=(threshold_value*2, 0.6),
            arrowprops=dict(facecolor='red', shrink=0.05, width=2), 
            fontsize=12, color='red', fontweight='bold')

ax.text(1000, 0.2, "Small Donations", fontsize=12, color='green', fontweight='bold')
ax.text(threshold_value*1.5, 0.9, "Large Donations", fontsize=12, color='darkred', fontweight='bold')

# Titles and labels
ax.set_title('Major Donors Drive Majority of Revenue', fontsize=22, fontweight='bold')
ax.set_xlabel('Donation Amount ($)', fontsize=16)
ax.set_ylabel('Cumulative % of Total Revenue', fontsize=16)

# Limits
ax.set_ylim(0, 1.05)
ax.set_xlim(0, 15000)  # Focus on relevant donation range

# Grid and background
ax.grid(True, linestyle='--', alpha=0.7)
fig.patch.set_facecolor('#f7f9fb')

# Format x-axis nicely
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x):,}'))

ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Set a sensible major donation threshold
threshold_value = 500  # You can change to 1000 if needed

# Group by large and small donations
large_donations = data[data['Revenue Amount'] > threshold_value]['Revenue Amount']
small_donations = data[data['Revenue Amount'] <= threshold_value]['Revenue Amount']

large_sum = large_donations.sum()
small_sum = small_donations.sum()
large_count = large_donations.count()
small_count = small_donations.count()

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar([f'Major Donations\n(> ${threshold_value:,})', 
               f'Smaller Donations\n( ${threshold_value:,})'], 
              [large_sum, small_sum], 
              color=['#fb8500', '#219ebc'],
              edgecolor='black',
              width=0.6)

# Annotate values above bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'${height/1e3:.1f}k', 
                xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 5),
                textcoords='offset points',
                ha='center', va='bottom',
                fontsize=13, fontweight='bold', color='black')

# Donor counts below bars
ax.text(bars[0].get_x() + bars[0].get_width()/2, -0.1*large_sum, 
        f'{large_count} donors', ha='center', fontsize=11, color='#fb8500', fontweight='bold')

ax.text(bars[1].get_x() + bars[1].get_width()/2, -0.1*small_sum, 
        f'{small_count} donors', ha='center', fontsize=11, color='#219ebc', fontweight='bold')

# Titles
ax.set_title('Impact of Major vs Smaller Donations\nMajor Donors Drive the Bulk of Revenue', 
             fontsize=18, fontweight='bold')

ax.set_ylabel('Total Donation Amount ($)', fontsize=14)
ax.set_ylim(bottom=min(0, small_sum * -0.2))

# Y-axis formatting
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}k'))

# Styling
ax.grid(axis='y', linestyle='--', alpha=0.7)
fig.patch.set_facecolor('#f7f9fb')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import make_interp_spline

# --- Prepare your data ---
# Example: replace this with your real dataset
# data = your actual dataframe
# Make sure 'Year' and 'Revenue Amount' columns exist
years = np.arange(2010, 2025)
revenues = [5000, 6000, 10000, 8000, 15000, 20000, 25000, 30000, 50000, 70000, 60000, 90000, 120000, 600000, 20000]
data = pd.DataFrame({'Year': years, 'Revenue Amount': revenues})

revenue_by_year = data.groupby('Year')['Revenue Amount'].sum()

# --- Create smooth curve ---
years_numeric = revenue_by_year.index.values
revenue_numeric = revenue_by_year.values

xnew = np.linspace(years_numeric.min(), years_numeric.max(), 300)
spl = make_interp_spline(years_numeric, revenue_numeric, k=3)  # cubic smoothing
y_smooth = spl(xnew)

# --- Plot ---
fig, ax = plt.subplots(figsize=(12, 7))

# Smooth line
ax.plot(xnew, y_smooth, color='darkgreen', linewidth=3)

# Original points
ax.scatter(years_numeric, revenue_numeric, color='darkgreen', edgecolor='black', s=70, zorder=5)

# Calculate YoY growth
growth = revenue_by_year.pct_change() * 100

# Label YoY growth
for year, value in growth.dropna().items():
    label = f"{value:+.1f}%"
    color = 'green' if value > 0 else 'red'
    offset = 10 if value > 0 else -15
    ax.annotate(label, 
                (year, revenue_by_year.loc[year]), 
                textcoords="offset points", 
                xytext=(0, offset), 
                ha='center', 
                fontsize=9, 
                color=color, 
                fontweight='bold')

# --- Highlight biggest jump and biggest drop ---
biggest_jump_year = growth.idxmax()
biggest_drop_year = growth.idxmin()

# Arrow for biggest jump
ax.annotate('Biggest Jump', 
            xy=(biggest_jump_year, revenue_by_year.loc[biggest_jump_year]), 
            xytext=(biggest_jump_year-1, revenue_by_year.loc[biggest_jump_year]*1.2),
            arrowprops=dict(facecolor='green', shrink=0.05, width=2, headwidth=8),
            fontsize=12, color='green', fontweight='bold')

# Arrow for biggest drop
ax.annotate('Biggest Drop', 
            xy=(biggest_drop_year, revenue_by_year.loc[biggest_drop_year]), 
            xytext=(biggest_drop_year-1, revenue_by_year.loc[biggest_drop_year]*0.5),
            arrowprops=dict(facecolor='red', shrink=0.05, width=2, headwidth=8),
            fontsize=12, color='red', fontweight='bold')

# --- Styling ---
ax.set_title(' Donation Growth Over Years\nSmoothed Curve with Key Highlights', fontsize=20, fontweight='bold', color='#2b2d42')
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Total Donations ($)', fontsize=14)

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}k'))
ax.grid(True, linestyle='--', alpha=0.6)
fig.patch.set_facecolor('#f9f9f9')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

# Step 2: Separate groups
recurring_donations = data[data['Donation Type'] == 'Recurring']['Revenue Amount']
one_time_donations = data[data['Donation Type'] == 'One-Time']['Revenue Amount']

# Step 3: Perform T-Test
t_stat, p_value = ttest_ind(recurring_donations, one_time_donations, equal_var=False)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print(" Recurring donations are significantly different from One-Time donations.")
else:
    print(" No significant difference between Recurring and One-Time donations.")
```

```python
# Step 1: Get each donor's first year of donation
first_donation_year = data.groupby('Account Number')['Year'].min()

# Step 2: Map donor type: New or Returning
data['Donor Type'] = data.apply(
    lambda row: 'New Donor' if row['Year'] == first_donation_year.get(row['Account Number']) else 'Returning Donor',
    axis=1
)

from scipy.stats import chi2_contingency

# Contingency Table
contingency = pd.crosstab(data['Donation Type'], data['Donor Type'])

# Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency)

print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print(" New vs Returning donors have a statistically significant difference in donation behavior.")
else:
    print(" No significant difference between New and Returning donors.")
```
